```python
import pandas as pd
df = pd.read_csv('Iris.csv')
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
Kmeans = KMeans(n_clusters = 3)
ac = AgglomerativeClustering(n_clusters=4)
```

```python
X = df.iloc[:,1:4]
print(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X.SepalLengthCm,X.SepalWidthCm,X.PetalLengthCm)

ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalWidthCm')
ax.set_zlabel('PetalLengthCm')
```

```python
Kmeans.fit(X)
ac.fit(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X.SepalLengthCm,X.SepalWidthCm,X.PetalLengthCm,c=Kmeans.labels_,cmap='rainbow')
ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalWidthCm')
ax.set_zlabel('PetalLengthCm')
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X.SepalLengthCm,X.SepalWidthCm,X.PetalLengthCm,c=ac.labels_,cmap='rainbow')
ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalWidthCm')
ax.set_zlabel('PetalLengthCm')
```
