```python
import numpy as np
def find_min(numbers):#find minnum
    if not numbers:  
        return None
    
    min_value = numbers[0]  
    for num in numbers:
        if num < min_value:
            min_value = num 
    
    return min_value


my_list = [5, 2, 9, 1, 5, 6]
min_value = find_min(my_list)
print("minnum:", min_value)
###############################################
def find_min_with_pop(numbers):#use pop
    if not numbers:  
        return None
    
    min_value = numbers.pop()  
    
    while numbers:
        num = numbers.pop()
        if num < min_value:
            min_value = num  
    
    return min_value


my_list = [5, 2, 9, 1, 5, 6]
min_value = find_min_with_pop(my_list.copy()) #use list.copy()
print("minnum:", min_value)
##################################################
def find_max(numbers):#find maxnum
    if not numbers:  
        return None
    
    max_value = numbers[0]  
    for num in numbers:
        if num > max_value:
            max_value = num 
    
    return max_value


my_list = [5, 2, 9, 1, 5, 6]
max_value = find_max(my_list)
print("maxnum:", max_value)
##################################################
def find_num_closest_ave(numbers):#find num closest to average
    if not numbers:  
        return None
    
    average_num = np.mean(numbers)
    
    closest_ave_num = numbers[0]  
    for num in numbers:
        if  abs(average_num-closest_ave_num ) > abs(average_num-num):
            closest_ave_num = num 
    
    return closest_ave_num


my_list = [5, 2, 9, 1, 5, 6]
closest_ave_num = find_num_closest_ave(my_list)
print("closest_ave_num:", closest_ave_num)
##################################################

def find_longest_number(numbers):#find_longest_number

    max_length = 0
    longest_numbers = []

    for num in numbers:
        str_num = str(num)
        length = len(str_num)

        if length > max_length:
            max_length = length
            longest_numbers = [num]
        elif length == max_length:
            longest_numbers.append(num)
            
    return longest_numbers

my_list = [123, 45, 7890, 12, 56789, 555, 9876]

longest_numbers = find_longest_number(my_list)
print("longest_numbers:",longest_numbers)
```
