```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 50)

y1 = x

y2 = [i**3 for i in x] # x^3

fig, ax = plt.subplots()
ax.set_title("y1 = x, y2 = x^3")
ax.grid()
ax.plot(x, y1,'--g', x, y2,'^k')
```
